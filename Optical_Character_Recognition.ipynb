```python
!pip install python-doctr[torch] # Install DocTR with PyTorch support
!pip install tf2onnx==1.13.0

from doctr.io import DocumentFile
from doctr.models import ocr_predictor

# Load the OCR model
model = ocr_predictor(pretrained=True)

# Function to perform OCR using DocTR
def perform_ocr(image_path):
    doc = DocumentFile.from_images(image_path)
    result = model(doc)

    # Extract the text
    extracted_text = result.render()
    return extracted_text

# File upload handling
from google.colab import files
from IPython.display import display, Javascript

# Create an upload button
display(Javascript('''
  const uploadButton = document.createElement('button');
  uploadButton.textContent = 'Choose Image';
  uploadButton.addEventListener('click', () => {
    google.colab.files.upload();
  });
  document.body.appendChild(uploadButton);
'''))

# Function to handle file upload and perform OCR
def upload_and_ocr():
    uploaded = files.upload()
    for fn in uploaded.keys():
        image_path = fn
        extracted_text = perform_ocr(image_path)
        print("Extracted Text:")
        print(extracted_text)

# Trigger the upload and OCR process
upload_and_ocr()
```
